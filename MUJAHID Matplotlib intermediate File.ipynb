```python
import numpy as np
import matplotlib.pyplot as plt
```

# Figure,axes&spines:

```python
fig, axs = plt.subplots(3,3)
axs[0,0].set_facecolor('#ddddff')
axs[2,2].set_facecolor('#ffffdd')
```

```python
gs =fig.add_gridspec(3,3)
ax =fig.add_subplot(gs[0, :])
ax.set_facecolor('#ddddff')
```

```python
fig, ax =plt.subplots()
ax.spines['top'].set_color('None')
ax.spines['right'].set_color('None')
```

```python
from mpl.ticker import MultipleLocator as ML
from mpl.ticker import Scalarformatter as SF
ax.xaxis.set_minor_locator(ML(0.2))
ax.xaxis.set_minor_formatter(SF())
ax.tick_params(axis='x',which='minor',rotation=90)
```

# Lines & Markers

```python
x = np.linspace(0.1, 10*np.pi, 1000)
y = np.sin(x)
plt.plot(x,y, 'C1o:', markevery=25, mec='1.0')
```

# Scales & Projections

```python
fig, ax= plt.subplots()
ax.set_xscale('log')
plt.plot(x,y,'C1o--', markevery=25,mec='1.0')
```

# Text & Ornaments

```python
plt.fill_betweenx([-1,1],[0],[2*np.pi])
plt.text(0,-1,r'Period $\Phi$')
```

# Legend

```python
plt.plot(x, np.sin(x),'C0', label='Sine')
plt.plot(x, np.cos(x),'C1', label='Cosine')
plt.legend(bbox_to_anchor=(0,1,1,.1), ncol=2, mode='expand',loc='lower left')
```

# Annotation

```python
plt.annotate('A', (x[250],y[250]),(x[250],-1),ha='center',va='center',arrowprops={'arrowstyle':'->','color':'C1'})
```
