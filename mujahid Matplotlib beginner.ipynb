# Organize

```python

y1,y2=np.sin(x), np.cos(x)
plt.plot(x,y1,x,y2)
```

```python
fig, (plt1, plt2) = plt.subplots(2,1)
plt1.plot(x, y1, color='C1')
plt2.plot(x, y2, color='C0')
```

```python
fig, (plt1, plt2) = plt.subplots(1,2)
plt1.plot(y1, x, color='C1')
plt2.plot(y2, x, color='C0')
```

# lable

```python
plt.plot(x,y)
fig.suptitle(None)
plt.title('A sine wave')
plt.xlabel('time')
plt.ylabel('Voltage')
```
